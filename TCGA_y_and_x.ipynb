```python
import pandas as pd

df_gene = pd.read_csv('genes_vitamind_only_transposed.csv', index_col=0)
df_gene.head()
```

**NORMALIZE GENE DATA AND MAKE INTO A SINGLE ROW**

```python
from sklearn.preprocessing import StandardScaler

selected_columns = df_gene.iloc[:, 1:] 
scaler=StandardScaler()
scaler.fit(selected_columns)
scaled_data=scaler.transform(selected_columns)
```

```python
selected_columns
```

```python
scaled_data
```

```python
df_gene2 = pd.DataFrame(scaled_data)
df_gene2
```

```python
df_gene2['mean'] = df_gene2.mean(axis=1)
df_gene2

#df['Y'] = df.iloc[:, 0] + df.iloc[:, 1] + df.iloc[:, 2] + df.iloc[:, 3] - df.iloc[:, 4] - df.iloc[:, 5]
```

```python
df_gene2['Y'] = df_gene2['mean']
df_gene2
```

```python
df_geneY = df_gene2[['Y']]
df_geneY
```

**NORMALIZE THE miRNA DATA**

```python
df_mirna = pd.read_csv("mirna_tpm_vals_jun22.csv", index_col=0)
print(df_mirna.shape)
df_mirna.head()
```

```python
column_names = df_mirna.columns[1:].tolist()
column_names
```

```python
from sklearn.preprocessing import StandardScaler

selected_columns_miRNA = df_mirna.iloc[:, 1:] 
scaler_miRNA=StandardScaler()
scaler_miRNA.fit(selected_columns_miRNA)
scaled_data_miRNA=scaler_miRNA.transform(selected_columns_miRNA)
```

```python
scaled_data_miRNA.shape
```

```python
df_mirna_scaled = pd.DataFrame(scaled_data_miRNA)
df_mirna_scaled
```

```python
df_mirna_scaled.columns = column_names
df_mirna_scaled
```

```python
concatenated_df = pd.concat([df_mirna_scaled, df_geneY], axis=1)
concatenated_df
```

**RANDOMIZE THE DATA**

```python
concatenated_df_shuffled = concatenated_df.sample(n=len(concatenated_df), random_state=1)
concatenated_df_shuffled
```

**RESET INDEX AND DROP EMPTY ROWS AND COLUMNS**

```python
concatenated_df_shuffled1 = concatenated_df_shuffled.reset_index(drop=True)
concatenated_df_shuffled1
```

```python
concatenated_df_shuffled2 = concatenated_df_shuffled1.dropna()
concatenated_df_shuffled2
```

**BUILD THE MODEL**

```python
X = concatenated_df_shuffled2.drop('Y', axis=1)
X
```

```python
Y = concatenated_df_shuffled2.Y
Y
```

```python
import numpy as np
offset = 1e-9  # Adjust the offset value as needed


concatenated_df_shuffled2['logTPM'] = np.log(concatenated_df_shuffled2['Y'] + offset)
```

```python
Y = concatenated_df_shuffled2.logTPM
Y
```

```python
#X = concatenated_df_shuffled2.drop('Y', axis=1)
X
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_test.shape, Y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100)
model.fit(X_train, Y_train)
r2 = model.score(X_test, Y_test)
r2
```
